### *** Useful Functions ***

```python
# function for calcutating the requrired statistics using 'groupby' and 'agg'
def calc_stats(group_col):
    stats = df.groupby(group_col).agg({
    'runs': 'sum',
    'wkts': 'sum',
    'runs_per_ball': 'mean',
    'econ': 'mean',
    '4s': 'sum',
    '6s': 'sum',
    'sr': 'mean'
    }).reset_index()
    return stats
```

### Task 1: Team Performance Analysis

```python
# Team-wise performance metrics
team_stats = calc_stats('team')

# Identify top-performing teams
top_teams = team_stats.sort_values(by='runs', ascending=False).head(5)

# Plotting team-wise performance
plt.figure(figsize=(12, 8))
sns.barplot(x='runs', y='team', data=top_teams, palette='viridis')
plt.title('Top Performing Teams - Runs Scored')
plt.xlabel('Total Runs')
plt.ylabel('Teams')
plt.savefig('Top Performing Teams - Runs Scored.png')
plt.show()
```

### Task 2: Player Performance Analysis

```python
# Individual player statistics
player_stats = calc_stats('player')

# Identify leading run-scorers and wicket-takers
top_batsmen = player_stats.sort_values(by='runs', ascending=False).head(5)
top_bowlers = player_stats.sort_values(by='wkts', ascending=False).head(5)

# Plotting leading run-scorers and wicket-takers
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
sns.barplot(x='runs', y='player', data=top_batsmen, palette='Blues_r')
plt.title('Leading Run-scorers')
plt.xlabel('Total Runs')
plt.ylabel('Players')

plt.subplot(1, 2, 2)
sns.barplot(x='wkts', y='player', data=top_bowlers, palette='Greens_r')
plt.title('Leading Wicket-takers')
plt.xlabel('Total Wickets')
plt.ylabel('Players')

plt.tight_layout()
plt.savefig('Leading Run-scorers & Leading Wicket-takers.png')
plt.show()
```

### Task 3: Opposition and Ground Analysis

```python
# Performance against different oppositions
opposition_stats = calc_stats('opposition')

# Performance across different grounds
ground_stats = calc_stats('ground')

# Plotting performance against different oppositions
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
sns.barplot(x='runs', y='opposition', data=opposition_stats, palette='YlOrBr')
plt.title('Team Performance Against Different Oppositions')
plt.xlabel('Total Runs')
plt.ylabel('Oppositions')

# Plotting performance across different grounds
plt.subplot(1, 2, 2)
sns.barplot(x='runs', y='ground', data=ground_stats, palette='pastel')
plt.title('Team Performance Across Different Grounds')
plt.xlabel('Total Runs')
plt.ylabel('Grounds')

plt.tight_layout()
plt.savefig('Team Performance Across Different Oppositions and Grounds.png');
plt.show()
```

### Task 4: Temporal Analysis

```python
# Convert 'start_date' to datetime format
df['start_date'] = pd.to_datetime(df['start_date'])

# Performance trends over time
temporal_stats = calc_stats('start_date')

# Plotting performance trends over time
plt.figure(figsize=(15, 8))
plt.subplot(2, 1, 1)
sns.lineplot(x='start_date', y='runs', data=temporal_stats, color='b')
plt.title('Team Performance Over Time - Runs Scored')
plt.xlabel('Start Date')
plt.ylabel('Total Runs')

plt.subplot(2, 1, 2)
sns.lineplot(x='start_date', y='wkts', data=temporal_stats, color='r')
plt.title('Team Performance Over Time - Wickets Taken')
plt.xlabel('Start Date')
plt.ylabel('Total Wickets')

plt.tight_layout()
plt.savefig('Team Performance Over Time - Runs Scored - Wickets Taken.png');
plt.show()
```
